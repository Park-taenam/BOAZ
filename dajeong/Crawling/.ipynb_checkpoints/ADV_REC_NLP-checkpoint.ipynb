#### (3) body_infos (고객프로필 정보 수정)

```python
i=0
gender, height, weight = [],[],[]
for i in range(len(body_infos)):
    if len(body_infos[i].split('\n'))==2:
        gender.append(body_infos[i].split('\n')[0].split(',')[0])
        height.append(body_infos[i].split('\n')[0].split(',')[1])
        weight.append(body_infos[i].split('\n')[0].split(',')[2])
    else:
        gender.append(np.nan)
        height.append(np.nan)
        weight.append(np.nan)
```

#### (4) goods_infos (고객구매 정보 수정)

```python
i=0
goods, getsize = [],[]
for i in range(len(goods_infos)):
    goods.append(goods_infos[i].split('\n')[0])
    getsize.append(goods_infos[i].split('\n')[1])
```
