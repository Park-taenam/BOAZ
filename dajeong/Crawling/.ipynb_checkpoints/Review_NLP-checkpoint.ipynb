### Dict 기반 감정분석

**단순 리뷰 전처리 시작**/
**처리해야할 것 같은 부분**
- 느낌표, 물음표 등의 특수문자 제거
- 띄어쓰기
- 할것같아용, 욯ㅎ, 좋아여 등 문법적 오류
- ㅎㅎㅎ, ㅠㅠㅠ, ㅋㅋㅋ 등의 단순 자음/모음 삭제
- 숫자 제거 여부
    - 163/50 → 키/몸무게 : 컬럼 존재
    - 26사이즈 샀는데 : 컬럼 존재
    - 26사이즈 사는게 좋을것 같아요 : 컬럼 존재 X
        - 실구매 사이즈는 25, 리뷰에는 숫자 26만 존재, 좋다라는 긍정 분류로 생각됨
        - 근데 이런 경우는 엄청 작지 않을까 → 일단 숫자 삭제하고 진행
- 근데, 그러나, 근디 등을 기준으로 문장 어감이 달라짐
- 길이가 1인 글자
    - 지우기엔 의미가 있을수도 ex) 키, 핏, 짱 등

**Tool**
- KoNLPy
    - 한국어 처리 패키지, 새롭게 만들어진 미 등록 단어 인식 정도가 떨어짐, 사전에 단어 등록
- SoyNlp
    - 사용자 사전 및 형태소 분석 없이 Cohension 기반으로 토큰화 기능 제공, 이걸로 New 단어 탐색
- PyKoSpacing
    - 띄어쓰기가 되어있지 않은 문장을 띄어쓰기 한 문장으로 변경
- 참고 링크
    - https://wikidocs.net/92961

```python
rv = df['content'].to_list()
rvv = ''
for i in range(len(rv)):
    rvv = rvv + ' '+ rv[i]
```

```python
rvv = re.sub(r'([^a-zA-Z가-힣])',' ',rvv) #특수문자, 숫자, 단순 자음모음 삭제
rvv = rvv.lower() #소문자 변환
units = ['mm','cm','kg','g'] #단위 제거
for unit in units: #띄어쓰기
    rvv = rvv.replace(unit, '')
rvv = spacing(rvv)
rvv
```
